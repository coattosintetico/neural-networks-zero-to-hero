## INTRO

We'll be following the approach introduced in [A Neural Probabilistic Language Model](https://www.jmlr.org/papers/volume3/bengio03a/bengio03a.pdf), in which instead of going from the purely statistical approach to predicting words / characters, they follow an embedding + MLP approach. It's not the first paper that introduced that idea, but it's one of the most cited and populars when it comes to it. 

In the paper they use **words instead of characters**, but we'll analogously apply this approach with characters.

As in the previous lesson, we'll build the mappings of character-index and index-character. This is the way of working with characters numerically.

````python
unique_characters: list[str] = sorted(list(set("".join(words))))
# character_to_index
ctoi = {c: i + 1 for i, c in enumerate(unique_characters)}
# for the special value "." (beginnings and ends of words) we'll assign the integer 0
ctoi["."] = 0
# and the reverse counterpart, to be able to get the corresponding character to a given index
itoc = {i: c for c, i in ctoi.items()}
````

````python
ctoi
````

````python
itoc
````

## Build the dataset

Now that we have the words, we have to build our dataset: the `X` and `Y` tensors that will make the input-output pairs, that will let us train out NN and get the proper weights.

````python
# context length: how many characters do we take to predict the next one?
block_size = 3
X, Y = [], []

for w in words[:4]:
    print(w)
    # first we fill the context with all zeros (the character ".")
    context = [0] * block_size

    # we iterate over each character in the given word
    for ch in w + ".":
        # we get the index of the character
        ix = ctoi[ch]
        # and given the existing context that we have (i.e. previous characters that we read in this word),
        # we use that pair context-character to create an X-Y value in our dataset (input-output)
        X.append(context)
        # (remember that context is a set of indexes as well)
        Y.append(ix)
        # for printing, we convert them into characters instead of indexes
        print("".join(itoc[i] for i in context), "--->", itoc[ix])
        # finally, we append this given character to the context and slice that context so that its size
        # is equal to the block size we specified
        context = context[1:] + [ix]

# and finally we  convert those lists into vectors, because that's what we use to work with matrices
X = torch.tensor(X)
Y = torch.tensor(Y)
````

So up until now it's exactly the same that we were doing before, but instead of using bigrams (`block_size = 2`) we're using sets of $x_i$ (that will predict an output $y_i$) of arbitrary length.

## Embedding

An embedding gives the NN a **compact** and **learnable** representation of each character, even being able to establish similarities between characters. With the one hot encoding approach we were just fixing the representation, making it arbitrarily bigger and avoiding any possible learnability there.

Furthermore, you're separating the two concerns that we were mixing in our previous approach with one hot encoded vectors: the embedding learns how to *represent* those characters, and the first layer learns what to *do* with those characters.

Also, we have to take into account that the dimension of the vectors that we'll input into the NN will determine directly the size of the first weight matrix (remember that the first layer's weights are $W \in \mathbb{R}^{n \times h}$, where $n$ is the input dimension and $h$ the number of neurons that we want for that layer), so if we use a smaller input vector (e.g. two dimensional embedding vs a $3 \text{ context length} \times 27 \text{ possible characters} = 81$ one hot encoded vector) we'll be doing less computation.

We initialize the embeddings from random numbers. We selected 2 to be the number of dimensions that we want.

````python
C = torch.randn((27, 2))
````

> Note: here he makes an argument that we could interpret this first *"take the value of C with index equivalent to that of our character"* (i.e. picking the embedding for each character) as a first layer to which we don't apply an activation function, making them linear. This happens because a one hot encoding vector times a matrix is just picking a specific row from that matrix, which is equivalent to indexing. This means that we have just moved backwards the one hot encoding process and the embedding could be seen as "another layer" which doesn't have an activation function.

## Constructing the hidden layer

Given that we have `3 embeddings x 2 dimensions = 6`, we have to initialize a matrix of weights of dimension `(6 x n)`, where `n` is the number of neurons we want in that first hidden layer. Just to recap: each neuron will have `6` weights associated with it. Let's pick 100, arbitrary decision.

````python
W1 = torch.randn((6, 100))
b1 = torch.randn(100)
````

We would like to do `emd @ W1 + b1`, but right now the shapes don't match:

````python
embedding.shape
````

````python
W1.shape
````

We want to "concatenate" the embeddings so that they are `(25 x 6)`. We can do this in several ways, for example using `torch.cat`:

````python
embedding
````

````python
# con esto pillamos:
# - todas las filas
# - solo la primera columna
# - el vector 2D entero de esa primera "celda" de la primera columna
embedding[:, 0, :]
````

````python
torch.cat([embedding[:, 0, :], embedding[:, 1, :], embedding[:, 2, :]], 1).shape
````

There's a more scalable solution: `unbind`.

````python
# this gives us exactly what we achieved before, because unbind returns the slices along a given
# dimension (like "splitting" a tensor into multiple other tensors)
torch.cat(torch.unbind(embedding, 1), 1).shape
````

But there's an even LEANER approach to this, and it makes sense to use it. 

Pytorch sees tensors as **one dimensional arrays** with some special properties that represent its size and other things (in order to know more about [Pytorch Internals](https://blog.ezyang.com/2019/05/pytorch-internals/), check that blog post from ezyang) (storage offset, strides and shapes). We can check it out with:

````python
a = torch.arange(18)
a
````

````python
a.shape
````

````python
a.view(2, 9)
````

````python
a.view(3, 3, 2)
````

````python
a.view(3, 6)
````

````python
a.view(3, 6).storage()
````

So whenever we are manipulating the view of a tensor, it's an extremely efficient operation because no memory for the values of the tensor is being modified (no copies, etc.).

This means we can do:

````python
embedding.view(25, 6)
````

Going back to the operation that we wanted before (`emd @ W1 + b1`), we can achieve that with:

````python
# using -1 in the first dimension will make pytorch guess what it should be so that the numbers match
h = embedding.view(-1, 6) @ W1 + b1
````

````python
h.shape
````

These are the **hidden states**, the activations of each neuron (100 neurons for each of the given input examples `X[i]`).

Also, we can review that the correct thing is happening with the hidden states:

````python
(embedding.view(-1, 6) @ W1).shape
````

````python
b1.shape
````

So pytorch will broadcast `b1` as a vector of:
```
25, 100
 1, 100
```
and it'll repeat each element in its first dimension 25 times. 

(This is done knowing about broadcasting rules, which I'm not an expert of and I don't know if I want to be).

## Implementing the output layer

Following the same logic as before (review the formal definition of NN if this is not clear), we need a layer of `(100 x n)`, where `n` is the number of neurons and thus the number of outputs. Given that we want this layer to be the final one, `n = 27`. 

````python
W2 = torch.randn((100, 27))
b2 = torch.randn(27)
````

````python
logits = h @ W2 + b2
````

````python
logits.shape
````

Logits is the name that we give the actual "raw output" of our final layer, upon which we'll apply softmax later on in order to get a proper mathematical interpretation out of it (probability distribution).

> *A note on logits and their interpretability:*
> 
> When we get the final result of our NN `(h @ W2 + b2)`, we get a 27 dimensional array of either positive, negative, big or small numbers. We can't make an interpretation out of them. That's why we run a **softmax** operation on them, because softmax enforces that distribution of numbers to behave like a distribution of probabilities (all between 0 and 1, and all sum 1).

````python
counts = logits.exp()
probs = counts / counts.sum(1, keepdims=True)
````

Note that we're calling that `counts` because given the mathematical interpretation that we're taking of the final output thanks to the softmax, the `exp` takes the same place structurally as the counts took earlier. 

We **can't** get to the same place as we did in the bigram lesson, in which by using the statistical approach and the NN approach we arrived at the same conclusion because by using embeddings (i.e. representing input characters by 2 dimensional vectors instead of one-hot encoded vectors of 27 dimensions) we're reducing the free parameters that we wrangle, so we can't achieve that perfect match, but we're getting the benefit of **generalization**, and unseen examples that behave similar to seen examples are going to get nice probability distributions, contrary to the statistical approach, in which we would be fucked for unseen examples. 

Trying to take the statistical approach when we have more possible contexts than training examples (e.g. we have ~200k characters and for a 5-gram (4 chars of context) we'd have $27^4 = 531,000$ possible contexts) is what is called the *curse of dimensionality* or *data sparsity*.
